```python
class Solution:
    def __init__(self, a: list[int], ans: list[int], work: list[int], grow_or_shrink: bool):
         ##YOU CAN HAVE ANY NUMBER OF DATA STRUCTURES HERE
        self.a = a
        self.ans = ans
        self.work = work
        self.grow_or_shrink = grow_or_shrink
        
    ## NOTHING CAN BE CHANGED BELOW
        self._alg()

        
        
  ##Implement your code BELOW
  ##You can have any number of private variables and functions
    def __len__(self):
        self._l = len(a)
        return self._l

    def _alg(self):
        self.work[0] = 1

        while len(self.a) > 1:
            for i in range(len(self.a) // 2 - 1, -1, -1):
                self.work[0] += 1
                if self.grow_or_shrink:
                    self.heapify(self.a, len(self.a), i, "min")
                else:
                    self.heapify(self.a, len(self.a), i, "max")

            self.swap(self.a, len(self.a), self.grow_or_shrink)

        self.ans[0] = self.a[0]

    def heapify(self, a, n: int, i: int, heap_type: str):
        left = 2 * i + 1
        right = 2 * i + 2

        if heap_type == "min":
            smallest = i
            if left < n and a[left] < a[smallest]:
                smallest = left
            if right < n and a[right] < a[smallest]:
                smallest = right

            if smallest != i:
                a[i], a[smallest] = a[smallest], a[i]
                self.heapify(a, n, smallest, heap_type)
        else:
            largest = i
            if left < n and a[left] > a[largest]:
                largest = left
            if right < n and a[right] > a[largest]:
                largest = right

            if largest != i:
                a[i], a[largest] = a[largest], a[i]
                self.heapify(a, n, largest, heap_type)

    def swap(self, a, n, increase):
        n = len(a)
        a[0], a[n - 1] = a[n - 1], a[0]

        self.heapify(a, n - 1, 0, "min" if increase else "max")

        if increase:
            a.append(a[n - 1] + 1 if a[n - 1] >= a[0] else a[0] + 1)
        else:
            a.append(a[n - 1] - a[0])

        a.pop(0)
        a.pop(n - 2)
```
